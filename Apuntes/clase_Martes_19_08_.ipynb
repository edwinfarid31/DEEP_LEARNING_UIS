```python
import numpy as np
import matplotlib.pyplot as plt

f  = lambda x: 0.1*np.power(x,4) - 4*np.power(x,2) - x
df = lambda x: 0.4*np.power(x,3) - 8*x - 1


alpha = 0.1
num_iter = 10
x = 0.1


for i in range(num_iter):
    x = x - alpha * df(x)

print("Mínimo local aproximado en x =", x)
print("Valor de f(x) en el mínimo =", f(x))


X = np.linspace(-6, 6, 400)
Y = f(X)
plt.plot(X, Y, label="f(x)")
plt.scatter(x, f(x), color="red", label="Mínimo encontrado")
plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()

```
