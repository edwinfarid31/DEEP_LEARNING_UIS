```python
import numpy as np 

def error_cuadratico(y1,y2):
    '''Esta funcion retorna el error cuadratico existente entre dos listas  con el mismo numero de elementos '''

    y1= y1.reshape(1,y1.size)
    y2= y2.reshape(1,y2.size)

    n1= y1.shape
    n2= y2.shape

    if n1[1] == n2[1]:
        error = y1-y2
        error_cuadratico= np.power(error,2)
        return error_cuadratico
    else:
        print("No se puede calcular el error cuadratico debido a la incompatibilidad de las dimenciones")

#prueba
y1 = np.array([1.5, 2.3, 0.8, 4.0])
y2 = np.array([1.2, 2.5, 1.0, 3.8])

errorCuadratico = error_cuadratico(y1,y2)
print(errorCuadratico)
```

```python
import numpy as np 
##preguntar 
def error_cuatratico_medio(y1,y2):
    "esta funcion toma dos listas y les calcula el error cuadratico medio"

    y1 = y1.reshape(1,y1.size)
    y2 = y2.reshape(1,y2.size)

    n1= y1.shape
    n2= y2.shape

    if n1[1] == n2[1]:
        error = y1-y2
        error_cuadratico= np.power(error,2)
        error_cuadratico_medio = np.sum(error_cuadratico)
        return error_cuadratico_medio
    else:
        print("No se puede calcular el error cuadratico debido a la incompatibilidad de las dimenciones")

#valores Prueba 

y_1 = np.array([1, 2, 3, 4])
y_hat_1 = np.array([2, 1, 3, 5])

print(error_cuatratico_medio(y_1,y_hat_1))
```

```python
import numpy as np 

def calculo_gradiente(y1,y2,x):
    """Esta funcion calcula el gradiente version vector y los parametros de entrada deben ser y1,y2 y x en el mismo orden """

    try:
        y1= y1.reshape(1,y1.size)
        y2= y2.reshape(1,y2.size)

        y1_dim= y1.shape
        y2_dim= y2.shape
        x_dim= x.shape

        if (y1_dim[1]==y2_dim[1]) and (x_dim[1]==y1_dim[1]):

            resta= y1-y2
            producto= resta*x 
            respuesta = np.sum(producto, axis=1)
            respuesta = respuesta.reshape(1,respuesta.size)
            return respuesta 
        else: 
            respuesta= "los arreglos no tienen los tamaños esperados"
            return respuesta 
    except:
        return "Ingreso un dato invalido por favor verifique los datos ingresados"

y1 =np.array([9, 8, 11])
y2 =np.array( [8.9, 8.1, 10.6])
x =np.array( [[0.5,0.7,0.2],
     [0.4,0.9,0.6]])

resultado = calculo_gradiente(y1,y2,x)
print(f"el resultado es {resultado}")
```
